```python
import json
import pandas as pd
```

```python
data = []

with open('短短匹配A类/train.txt', encoding='utf-8')as f:
    for line in f.readlines():
        data.append(json.loads(line))
        
with open('短短匹配A类/valid.txt', encoding='utf-8')as f:
    for line in f.readlines():
        data.append(json.loads(line))  
        
with open('短短匹配B类/train.txt', encoding='utf-8')as f:
    for line in f.readlines():
        data.append(json.loads(line))
        
with open('短短匹配B类/valid.txt', encoding='utf-8')as f:
    for line in f.readlines():
        data.append(json.loads(line))
        
df1 = pd.DataFrame(data)
df1['type'] = 'ss'
df1['len1'] = df1['source'].apply(len)
df1['len2'] = df1['target'].apply(len)

print(df1.len1.describe(), df1.len2.describe())
df1
```
